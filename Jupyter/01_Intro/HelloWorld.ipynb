In diesme interaktiven Notebook werden wir die wichtigsten Bibliotheken **NumPy** und **Matplotlib** nutzen, um eine Funktion grafisch darzustellen. Zuallererst benötigen wir eine Funktion:

$$ y = \sin{x} + \cos{x} $$

Bevor wir die Bibliotheken nutzen, müssen wir sie zuerst laden:

```python
import numpy
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

Aus der Bibliothek **numpy** verwenden wir die Funktion `numpy.linspace()`, um uns einen Vektor mir gleichmäßig verteilten Zahlen zu erzeugen:

```python
x = numpy.linspace(0, 10*numpy.pi, 10**3)
y = numpy.sin(x) + numpy.cos(x)
```

Nun können wir unsere Funktion grafisch darstellen:

```python
plt.plot(x,y)
plt.title('Coole Funktion $y = \sin{x} + \cos{x} $');
```
